## Setup  
Load dependencies and set up environment for discussion 2.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

sns.set_theme(style="whitegrid", palette="muted")
```

## Data Acquisition  
Load and parse the TCGA Kidney Cancers RNA Seq Dataset. Separate the 'target' column (column of interest).

```python
print("Loading professor's dataset...")
df = pd.read_csv("kidney_tcga_dataset.csv")

target_col = 'target'

X = df.drop(columns=['sampleID', 'gender', 'sample_type', 'primary_site', target_col])
y = df[target_col]

print(f"Dataset successfully split!")
print(f"Features (X): {X.shape[1]} genes")
print(f"Samples: {X.shape[0]} patients")
print(f"Target classes to predict: {y.unique()}")
```

## Data Preprocessing & Splitting  
Split the data into training and testing sets (80/20). Prevents data leakage from PCA before spliting. Also, standardize the data to ensure all 60,000 genes are on the same scale.

```python
print("\n--- DATA SUMMARIZATION ---")
print("Target Class Distribution:")
print(y.value_counts(normalize=True).round(3) * 100)

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\nData successfully split and scaled.")
```

## Dimensionality Reduction  
Use PCA to compress the 60,498 genes to 50 principal, key components.

```python
print("\n--- DIMENSIONALITY REDUCTION (PCA) ---")

pca = PCA(n_components=50, random_state=42)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

plt.figure(figsize=(8, 6))

subtypes_for_plot = label_encoder.inverse_transform(y_train)

sns.scatterplot(
    x=X_train_pca[:, 0], 
    y=X_train_pca[:, 1], 
    hue=subtypes_for_plot, 
    palette='viridis', 
    alpha=0.8
)

plt.title("PCA Embedding of Kidney Cancer RNA-Seq Data (2D)")
plt.xlabel("Principal Component 1")

plt.ylabel("Principal Component 2")
plt.legend(title='Cancer Subtype')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 4))
sns.histplot(x=X_train_pca[:, 0], hue=subtypes_for_plot, kde=True, palette='viridis')
plt.title("Density Distribution of Principal Component 1")
plt.xlabel("Principal Component 1 Value")
plt.show()
```
